```python
#import required libraries
import random
import pandas as pd
import numpy as np
```

```python
# create dataset of 200 points
main_dataset = []
for x in range(200):
    main_dataset.append(random.randint(1,1000))
```

```python
# print the head value of datset
main_dataset = pd.DataFrame(main_dataset)
print(main_dataset[0].head())
```

```python
# Insert NaN values in the dataset
for i in range(0,20):
    x = random.randint(0,200)
    dirty_dataset_1 = main_dataset.replace(to_replace = main_dataset[0][x], value = np.NaN)
    print(dirty_dataset_1[0][x])
```

```python
# Make copy of dirty datset
DB_A = dirty_dataset_1.copy()
```

```python
# drop the rows having NaN
DB_A.dropna(how='any')
```

```python
# Replace NaN with the global constant
DB_B = dirty_dataset_1.copy()
global_constant = 20
DB_B[0].fillna(global_constant, inplace=False)
print(DB_B.head())
```

```python
# Replace NaN with mean value
DB_C = dirty_dataset_1.copy()
mean = np.mean(DB_C[0])
print(mean)
DB_C[0].fillna(mean, inplace=False)
print(DB_C.head())
```

```python
# Replace NaN with median value
DB_D = DB_B.copy()
median = np.median(DB_B[0])
print(median)
DB_D[0].fillna(median, inplace=True)
print(DB_D.head())
```

```python
#Create samples and find mean
list_mean=[]
DB_E = DB_B.copy()
# For 10 samples
for i in range(0,10):
    no_repl = DB_E.sample(n=20, replace= False)
    no_repl_mean = np.mean(no_repl)
    list_mean.append(no_repl_mean)
print(list_mean)
final_mean=np.mean(list_mean)
print("The final mean is")
print(final_mean)
```

```python
# Replace NaN with final mean
DB_E = dirty_dataset_1.copy()
DB_E[0].fillna(final_mean, inplace=True)
print(DB_E.head())
```

```python
# find standard deviation of each sample
main_std=np.std(main_dataset)
dirty_datset1_std=np.std(dirty_dataset_1)
DB_A_std=np.std(DB_A)
DB_C_std=np.std(DB_C)
DB_D_std=np.std(DB_D)
DB_E_std=np.std(DB_E)
```

```python
# print the standard deviation values
print(main_std, dirty_datset1_std, DB_A_std, DB_C_std, DB_D_std, DB_E_std)
```
