```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read url
data = pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
data.describe()
```

```python
data.drop(columns=['Channel','Region'])
```

```python
import pylab as pl
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=2).fit(data)
```

```python
kmeans.cluster_centers_
```

```python
y_pred = kmeans.predict(data)
```

```python
cluster_map = pd.DataFrame()
cluster_map['data_index'] = data.index.values
cluster_map['cluster'] = kmeans.labels_


for i in range (0,5):
    cnt=cluster_map[cluster_map.cluster == i].count()
    print(cnt)
```

```python
# Print elbow graph for the cluster
for x in range(0,3):
    Nc = range(2, 20)

    kmeans = [KMeans(n_clusters=i) for i in Nc]

    kmeans

    score = [kmeans[i].fit(data).score(data) for i in range(len(kmeans))]

    score

pl.plot(Nc,score)

pl.xlabel('Number of Clusters')

pl.ylabel('Score')

pl.title('Elbow Curve')

pl.show()
```

```python
# Calculate the sse for individual cluster(SSE)
sse = {}
for k in range(1, 100):
    kmeans = KMeans(n_clusters=k, max_iter=100).fit(data)
    data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
# print(sse)
plt.plot(sse.keys(), sse.values())
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```
