```python
# Function to calculate IQR
def get_subset_by_IQR(df,column):
    q1 = df[column].quantile(0.30)
    q3 = df[column].quantile(0.70)
    iqr = (df[column] >= q1) & (df[column] <= q3)
    print(q3-q1)
    return df.loc[iqr]
```

```python
# Find IQR of numerical features
get_subset_by_IQR(df,'age')
get_subset_by_IQR(df,'bmi')
get_subset_by_IQR(df,'children')
get_subset_by_IQR(df,'expenses')
```

```python
# Numerical features
numerical_features = ['age', 'bmi', 'children', 'expenses']

for col_name in numerical_features:
        # get iqr for each column
        iqr = get_subset_by_IQR(df, col_name)
        # find boxplot for iqr of col_name
        boxplot = iqr.boxplot(column = col_name)
        plt.show()
```
